```python
import plotly.graph_objects as go
import numpy as np

# 'un2022_03' 값에 따라 그룹을 나누기 위한 인덱스 딕셔너리
group_indices = {'X': 1, 'J': 2, 'O': 3}

# Initialize empty lists for each group type
x_values = []
y_values = []
size_values = []
colors = []  # 그룹별로 색상을 지정하기 위한 리스트

# Separate data based on 'un2022_03' value
for item in data:
    if item['un2022_03'] == 'X':
        group = 'X'
        angle = group_indices[group] * 120 + np.random.uniform(-20, 20)  # Adjust angle for 'X' group
        colors.append('red')  # 'X' 그룹은 빨간색으로 지정
    elif item['un2022_03'] == 'J':
        group = 'J'
        angle = group_indices[group] * 120 + np.random.uniform(-15, 15)  # Adjust angle for 'J' group
        colors.append('#30c67c')  # 'J' 그룹은 초록색으로 지정
    else:
        group = 'O'
        angle = group_indices[group] * 120 + np.random.uniform(-30, 30)
        colors.append('#40c9ff')  # 'O' 그룹은 파란색으로 지정

    radius = np.random.uniform(0.6, 0.9)  # Adjust the radius for each bubble

    x_values.append(radius * np.cos(np.deg2rad(angle)))
    y_values.append(radius * np.sin(np.deg2rad(angle)))
    size_values.append(item['power'])

# Scale size_values using square root to amplify differences
scaled_size_values = np.sqrt(size_values)

# Normalize scaled_size_values within the entire dataset to the range [0.3, 0.8]
max_size = max(scaled_size_values)
min_size = min(scaled_size_values)
normalized_size = [(size - min_size) / (max_size - min_size) * 1 + 0.3 for size in scaled_size_values]

# Adjust bubble size for 'X' group to make it larger
for i in range(len(data)):
    if data[i]['un2022_03'] == 'X':
        normalized_size[i] *= 1.2

# 그룹별 범례 라벨 설정
group_legends = {'O': '찬성', 'X': '반대', 'J': '중립'}

# 버블 차트 생성
fig = go.Figure()

# Add trace for each group separately
fig.add_trace(go.Scatter(
    x=x_values,
    y=y_values,
    mode='markers',
    marker=dict(
        size=[normalized_size[i] * 30 for i in range(len(normalized_size))],  # 버블 크기를 일정 비율로 조정
        sizemode='diameter',
        opacity=0.7,
        color=colors,  # 그룹별로 색상 지정
    ),
    name='',  # 'Bubbles' 범례 제거
    text=[item['country'] for item in data],
))

# 레이아웃 설정
fig.update_layout(
    title='우크라이나 침공 반대 유엔결의안',
    showlegend=True,
    xaxis=dict(
        showgrid=False,
        zeroline=False,
        showticklabels=False,
        showline=False,  # x축 라인 제거
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showticklabels=False,
        showline=False,  # y축 라인 제거
    ),
    plot_bgcolor='white',  # 배경색을 하얀색으로 설정
    font=dict(size=18, color='rgb (47,79,79)', family='Noto Sans KR, bold' )
)

# 그룹별 범례 추가
for group, legend_label in group_legends.items():
    fig.add_trace(go.Scatter(
        x=[None],
        y=[None],
        mode='markers',
        marker=dict(size=10, color='red' if group == 'X' else '#30c67c' if group == 'J' else '#40c9ff'),
        showlegend=True,
        name=legend_label,
    ))


fig.show()
```

```python
# '제재' 열 값이 1인 데이터만 가져옴
sanctioned_countries = df[df['제재'] == 1]

# 차트 생성
fig = go.Figure()

# 큰 원의 중심 좌표
center_x = 0
center_y = 0

# 큰 원의 초기 반지름
big_circle_radius = 500

# 원들을 배치하기 위한 각도들 (6개의 균등한 각도 생성)
angles = np.linspace(0, 2 * np.pi, len(sanctioned_countries), endpoint=False)

# 다른 원들 추가 (좌표 범위 내에서 랜덤하게 설정)
for i, country in enumerate(sanctioned_countries['country']):
    # 국가별 인구수에 비례하여 원의 크기 조정
    population = sanctioned_countries.iloc[i]['population']

    # 각도에 따라 원의 위치 계산
    country_un = center_x + big_circle_radius * np.cos(angles[i])
    country_power = center_y + big_circle_radius * np.sin(angles[i])

    # 대시라인에 연결되도록 설정
    fig.add_trace(go.Scatter(x=[country_un, 0],
                             y=[country_power, 0],
                             mode='lines',  # 선만 사용
                             line=dict(color='rgba(0, 0, 255, 0.2)', dash='dash', width=1),  # 대시라인 설정 (투명도를 가진 파란색 대시라인)
                             text=country,  # 대시라인 위에 표시할 텍스트
                             hoverinfo='text'
                             ))

    fig.add_trace(go.Scatter(x=[country_un],
                             y=[country_power],
                             mode='markers+text',  # 마커(버블 모양)와 텍스트 함께 사용
                             marker=dict(size=np.sqrt(population) / 250, symbol='circle', color='#40c9ff'),  # 원의 크기와 색상 설정
                             text=country,  # 원 위에 표시할 텍스트
                             hoverinfo='text',
                             textposition='middle center'  # 텍스트 위치 설정 (원의 상단 중앙에 표시)
                             ))

    # 다음 원의 반지름을 변경 (500 -> 400 또는 400 -> 500)
    big_circle_radius = 850 - big_circle_radius  # 900은 500과 400의 합

# 러시아 원을 덮어서 가장 위에 표시되도록 설정
fig.add_trace(go.Scatter(x=[0], y=[0], mode='markers', marker=dict(size=80, color='red', symbol='circle'),
                         name='러시아', text='러시아', hoverinfo='text', textposition='middle center'))

# 범례 추가: 원들의 개수를 나타내는 범례
fig.add_trace(go.Scatter(x=[], y=[], mode='markers', marker=dict(size=10, color='blue', symbol='circle'), showlegend=True, name=f'country 개수: {len(sanctioned_countries)}'))

# 하단에 country 수를 표시하는 텍스트 어노테이션 추가
fig.add_annotation(
    go.layout.Annotation(
        x=0,
        y=-700,
        text=f"수출 제재 국가들: {len(sanctioned_countries)} 개국",
        showarrow=False,
        font=dict(size=12, color='black'),
    )
)

# 레이아웃 설정 (범례 포함)
fig.update_layout(
    title='러시아 제재',
    showlegend=False,  # 범례 추가
    plot_bgcolor='white',  # 레이아웃 배경색을 흰색으로 설정
    xaxis=dict(visible=False),  # x축 값 숨기기
    yaxis=dict(visible=False),  # y축 값 숨기기
    font=dict(size=18, color='rgb (47,79,79)', family='Noto Sans KR, bold')
)

# 차트 출력
fig.show()
```
