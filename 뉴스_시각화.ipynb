import pandas as pd
from keybert import KeyBERT
import dask.dataframe as dd
from dask.distributed import Client

# CSV 파일을 Dask DataFrame으로 읽습니다.
df = dd.read_csv('./202305까지.CSV', encoding='utf-8', parse_dates=['일자'])

# Dask 클라이언트를 초기화합니다.
client = Client()

def extract_top_keywords_by_month(df):
    # '일자' 열에서 년도와 월을 추출하여 새로운 열 'YearMonth'에 추가합니다.
    df['YearMonth'] = df['일자'].dt.to_period('M')

    # KeyBERT 인스턴스를 생성합니다.
    kw_extractor = KeyBERT('distilbert-base-nli-mean-tokens')

    def process_group(group_df):
        # 현재 월의 모든 텍스트를 연결합니다.
        text_concatenated = ' '.join(group_df['키워드'])

        # 그룹 크기를 얻습니다.
        group_size = len(group_df)

        # 그룹 크기에 따라 상위 키워드 추출 개수를 결정합니다.
        num_top_keywords = min(2, group_size)

        # KeyBERT를 사용하여 키워드를 추출합니다.
        keywords = kw_extractor.extract_keywords(text_concatenated)

        # 월별 상위 키워드만 선택합니다.
        top_keywords = keywords[:num_top_keywords]

        # 월별 상위 키워드를 저장할 DataFrame을 생성합니다.
        df_result = pd.DataFrame({
            'YearMonth': [group_df['YearMonth'].iloc[0]],
            'Keyword': [kw[0] for kw in top_keywords],
            'Weight': [kw[1] for kw in top_keywords]
        })

        # 월별 상위 키워드가 담긴 DataFrame을 반환합니다.
        return df_result

    # 'YearMonth'로 데이터를 그룹화하고, 각 그룹에 process_group 함수를 적용합니다.
    top_keywords_by_month = df.groupby('YearMonth').apply(process_group, meta=pd.DataFrame(columns=['YearMonth', 'Keyword', 'Weight']))

    return top_keywords_by_month

# 함수를 호출하여 각 월별 상위 키워드를 추출합니다.
result = extract_top_keywords_by_month(df)

# 결과를 출력합니다.
print(result.compute())
